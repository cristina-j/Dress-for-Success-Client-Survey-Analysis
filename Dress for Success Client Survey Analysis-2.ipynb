# Dress for Success Client Survey Analysis

Dress for Success is a non-profit organization dedicated to empowering women to achieve economic independence by providing professional attire, career development tools, and ongoing support. The organization opperated across multiple locations in North Carolina, offerring both in-person and virtual prorgams that help women prepare for job interviews, build resume, and strengthen career-readiness skills. 

Overtime, Dress for Success noticed variations in client engagement across locations and services. To optimize program delivery and ensure quitable access, they sought a deeper understanding of client service usage, outcomes, and feedback patterns with a survey form they created to learn more about their clients. 

## Business Problem 

Dress for Success wants to analyze the responses across key focus areas to examine the outcomes by service location and explore regional differences in client experience, access, and results.

The organization wanted to answer a few questions to improve their services and client experience:

- Which service location(s) are most utilized?
- Which services were accessed remotely vs. in-person?
- Is there a pattern in outcomes by location?
- What services or programs are missing from the client perspective?
- What feedback have clients shared?

## Proposed Solution

Using anonymized client survey data, I performed a data cleaning, analysis, and reporting process using Microsoft Excel and pivot table dashboards.

The analysis focused on:
- Standardizing and validating location and service data
- Summarizing service usage frequencies and delivery methods
- Comparing in-person vs. virtual engagement
- Aggregating feedback responses to identify satisfaction patterns
- Visualizing results through charts and dashboards for board-level reporting

All findings were aggregated and anonymized to maintain data confidentiality.

# Geographical and Delivery Channel Analysis

This section will be a summary of my findings for each question the business wanted answers to and what these findings mean.

## Question #1: Which service location(s) are most utilized?

Analysis of client participation across Dress for Success Triangle service locations revealed that Raleigh is the most active site, with 123 participants (36%) reporting that they accessed services there.

Virtual services ranked second, accounting for 84 participants (25%), reflecting the organization’s growing success in expanding digital accessibility.

The Durham location and Mobile Boutique each served 58 participants (17%), indicating comparable engagement levels between in-person and outreach-based programming.

This distribution highlights both the centrality of the Raleigh office and the strong adoption of virtual services which is a key opportunity for expanding reach beyond physical locations.

## Question #2: Which services were accessed remotely vs. in-person?

The results show the number of unique participants who received services remotely.

When looking at how clients accessed services revealed that Career Coaching was the most frequently accessed service remotely, with 33 participants (23.7%) reporting they received this service online. This indicates that Career Coaching effectively adapts to virtual delivery, making it accessible for clients facing barriers where they are not able to recieve these services online.

Career Pathways, on the other hand, had only one participant (0.7%) accessing the service remotely. This lower rate is expected since Career Pathways is a structured, annual program that primarily occurs in person.


## Question #3: Is there a pattern in outcomes by location?

When we compare service use to skills that were gained, not every service led to a skill. However, many people are gaining skills, especially in Raleigh and remotely.

Across all the skills gained, approximately 87% of people have stated they have gained a skill from the Raleigh location.

Skills gained remotely comes in second at 51%.

When looking at the unique number of participants (individuals who have not been counted twice and have chosen at least one skill), Raleigh is still at the top at 44%.

# Client Voice and Qualitative Analysis

## Questions #1: What services or programs are missing from the client perspective?

Among the 36 participants who offered concrete suggestions, Networking Events, which includes job fairs and hiring events spoke for 14% of the participants.

Financial Assistance includes housing and help with transportation so participants could go to interviews and work speaks for 17% of the participants.

Interviews includes mock interviews which speaks for 6% of the participants.

## Question #2: What feedback have client shared?

Client feedback revealed an overwhelmingly positive experience with Dress for Success Triangle’s services and staff.

More than half of participants shared reflections expressing gratitude, confidence gains, and appreciation for the impact the organization has had on their career journeys.

Many clients specifically highlighted that having access to both in-person and virtual services made a meaningful difference, allowing them to participate despite scheduling, transportation, or childcare challenges.

Several participants also mentioned staff members by name, emphasizing how the team’s encouragement, professionalism, and personal support helped them feel more confident and prepared for employment opportunities.

Overall, client sentiment reflects that Dress for Success Triangle’s combination of accessibility, compassionate staff, and high-quality services directly contributes to improved confidence and job outcomes.
